```python
X = df.drop(columns=['Id', 'SalePrice'])
Y = df[['SalePrice']]
```

```python
X.head()
```

```python
# Seperate the categorical and continuous features
cat = list(X.columns[X.dtypes=='object'])
con = list(X.columns[X.dtypes!='object'])
```

```python
cat
```

```python
con
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='mean')),
                           ('scaler', StandardScaler())])
```

```python
cat_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='most_frequent')),
                           ('ohe', OneHotEncoder(handle_unknown='ignore',
                                                 sparse_output=False))])
```

```python
pre = ColumnTransformer([('num', num_pipe, con),
                         ('cat', cat_pipe, cat)]).set_output(transform='pandas')
```

```python
X_pre = pre.fit_transform(X)
X_pre.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.2, random_state=42)

print("First few rows of xtrain:\n", xtrain.head())
print("First few rows of ytrain:\n", ytrain.head())
print("First few rows of xtest:\n", xtest.head())
print("First few rows of ytest:\n", ytest.head())
print("xtrain shape:", xtrain.shape)
print("xtest shape:", xtest.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import SequentialFeatureSelector
```

```python
lr = LinearRegression()
sel = SequentialFeatureSelector(lr, 
                                n_features_to_select='auto', 
                                direction='forward', n_jobs=-1)
sel.fit(X_pre, Y)
sel_cols = sel.get_feature_names_out()
print(sel_cols)
```

```python
len(sel_cols)
```

```python
imp_cols = [i.split('__')[1] for i in sel_cols]
imp_cols
```

```python
X_sel = X_pre[sel_cols]
X_sel.head()
```
