```python
X = df.drop(columns=['Id', 'SalePrice'])
Y = df[['SalePrice']]
```

```python
X.head()
```

```python
# Seperate the categorical and continuous features
cat = list(X.columns[X.dtypes=='object'])
con = list(X.columns[X.dtypes!='object'])
```

```python
cat
```

```python
con
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OrdinalEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipe1 = Pipeline(steps=[('impute', SimpleImputer(strategy='mean')),
                           ('scaler', StandardScaler())])
```

```python

cat_pipe1 = Pipeline(steps=[('impute', SimpleImputer(strategy='constant',fill_value='N')),
                           ('ordinal', OrdinalEncoder())])
```

```python
pre1 = ColumnTransformer([('num', num_pipe1, con),
                         ('cat', cat_pipe1, cat)]).set_output(transform='pandas')
```

```python
X_pre = pre1.fit_transform(X)
X_pre.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import SequentialFeatureSelector
```

```python
lr = LinearRegression()
sel = SequentialFeatureSelector(lr, 
                                n_features_to_select='auto', 
                                direction='backward')
sel.fit(X_pre, Y)
sel_cols = sel.get_feature_names_out()
print(sel_cols)
```

```python
len(sel_cols)
```

```python
imp_cols = [i.split('__')[1] for i in sel_cols]
imp_cols
```

```python
X_sel = X[imp_cols]
X_sel.head()
```

```python
cat_sel = list(X_sel.columns[X_sel.dtypes=='object'])
con_sel = list(X_sel.columns[X_sel.dtypes!='object'])
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
num_pipe2 = Pipeline(steps=[('impute', SimpleImputer(strategy='mean')),
                           ('scaler', StandardScaler())])
```

```python
cat_pipe2 = Pipeline(steps=[('impute', SimpleImputer(strategy='constant',fill_value='Not_Avail')),
                           ('ohe', OneHotEncoder(handle_unknown='ignore',
                                                 sparse_output=False))])
```

```python
pre2 = ColumnTransformer([('num', num_pipe2, con_sel),
                         ('cat', cat_pipe2, cat_sel)]).set_output(transform='pandas')
```

```python
X_sel_pre = pre2.fit_transform(X_sel)
X_sel_pre.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_sel_pre, Y,
                                                test_size=0.2, random_state=21)
```

```python
import numpy as np 
params1 = {'alpha':np.arange(0.1, 100, 0.1)}
params1
```

```python
params2 = {'alpha': np.arange(190, 210, 0.1)}
params2
```

```python
from sklearn.linear_model import Lasso 
model_lasso = Lasso()
gscv_lasso = GridSearchCV(model_lasso, param_grid=params2, cv=5, scoring='r2')
gscv_lasso.fit(xtrain, ytrain)
```

```python
gscv_lasso.best_params_
```

```python
gscv_lasso.best_score_
```

```python
best_lasso = gscv_lasso.best_estimator_
best_lasso
```

```python
best_lasso.score(xtrain, ytrain)
```

```python
best_lasso.score(xtest, ytest)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, \
                            mean_absolute_percentage_error, r2_score
```

```python
def evaluate(model, x, y):
    ypred = model.predict(x)
    mse = mean_squared_error(y, ypred)
    rmse = mse**(1/2)
    mae = mean_absolute_error(y, ypred)
    mape = mean_absolute_percentage_error(y, ypred)
    r2 = r2_score(y, ypred)
    print(f'Mean Squared Error : {mse:.2f}')
    print(f'Root Mean Squared Error : {rmse:.2f}')
    print(f'Mean Absolute error : {mae:.2f}')
    print(f'Mean Absolute Percentage Error : {mape:.4f}')
    print(f'R2 score : {r2:.4f}')
```

```python

# Linear Regression
print("Linear Regression\n")
print("Train Results : ")
evaluate(model1, xtrain, ytrain)
print('\n-----------------------\n')
print("Test Results : ")
evaluate(model1, xtest, ytest)
```

```python
# Ridge
print("Ridge\n")
print("Train Results : ")
evaluate(best_ridge, xtrain, ytrain)
print('\n-----------------\n')
print("Test Results : ")
evaluate(best_ridge, xtest, ytest)
```

```python
# Lasso
print("Lasso\n")
print("Train Results : ")
evaluate(best_lasso, xtrain, ytrain)
print('\n------------------------\n')
print("Test Results : ")
evaluate(best_lasso, xtest, ytest)
```
